```python
sales_corr = sales_db.select_dtypes(include=['int','float']).corr()
sales_corr
```

```python
# Identify and exclude non-numeric columns
numeric_columns = sales_db.select_dtypes(include=['int', 'float'])

# Compute correlation matrix
sales_corr = numeric_columns.corr()

# Display correlation matrix
print(sales_corr)
```

```python
fig = plt.figure(figsize = (8,8))
plt.matshow(sales_corr, cmap = 'RdBu', fignum=fig.number)
plt.xticks(range(len(sales_corr.columns)), sales_corr.columns, rotation='vertical')
plt.yticks(range(len(sales_corr.columns)), sales_corr.columns)
```
